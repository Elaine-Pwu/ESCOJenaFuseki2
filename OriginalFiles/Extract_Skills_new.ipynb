```python
from SPARQLWrapper import SPARQLWrapper, JSON
import pandas as pd

sparql = SPARQLWrapper("https://ec.europa.eu/esco/sparql/esco-v1.2.0/query")

sparql.setQuery("""
PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
PREFIX esco: <http://data.europa.eu/esco/model#>

SELECT ?skill ?label
WHERE {
    ?skill a esco:Skill ;
           skos:prefLabel ?label .
    FILTER(lang(?label)='en')
}
LIMIT 100
""")

sparql.setReturnFormat(JSON)
# try :
#     sparql.setReturnFormat(JSON)
#     ret = sparql.query()
#     d = ret.convert()
# except Exception as e:
#     print(e)
results2 = sparql.query().convert()
print(results2)
# df = results2.query_to_dataframe()
df = pd.json_normalize(results2["results"]["bindings"])

# Clean columns (remove 'value')
df = df.map(lambda x: x.get("value") if isinstance(x, dict) else x)
df = df.rename(columns={"skill.value": "skill", "label.value": "label"}) 
df[["skill", "label"]]
# df
# dict
# print(df)
# results2
```

```python
sparql = SPARQLWrapper("https://ec.europa.eu/esco/sparql/esco-v1.2.0/query")

from SPARQLWrapper import CSV


sparql.setQuery ("""" 
    PREFIX skos: <http://www.w3.org/2004/02/skos/core#>
    PREFIX esco: <http://data.europa.eu/esco/model#>
    SELECT ?skill ?label 
    WHERE {
        ?skill a esco:Skill; skos:prefLabel ?label.
        FILTER(lang(?label) = "en")
    }
""")
skills = sparql.queryAndConvert()
print(skills)
print(type(skills))

skill_list = []
for skill in skills:
    # toPython --> convert Literal to only the text part we want
    skill_list.append(skill.label.toPython())
print(len(skill_list))
print(skill_list[:10])

df_skills = pd.DataFrame({"skill": skill_list})
df_skills.to_csv("skill_df.csv", encoding = "utf8")
```
